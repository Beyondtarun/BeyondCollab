```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/vivekpenuli/Ml_class/main/Kmean/Mall_Customers.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,[3,4]].values
```

```python
# choose the number of clusters- elbow test ->wcss (inertia)
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (1,11):
  ob = KMeans(n_clusters=i)
  ob.fit(x)  #kMeans model applied fr training
  wcss.append(ob.inertia_) #calculate distance wcss
```

```python
plt.plot(range(1,11),wcss)
plt.title("Elbow test")
plt.xlabel("No. of cluster")
plt.ylabel("wcss")
plt.show()
```
